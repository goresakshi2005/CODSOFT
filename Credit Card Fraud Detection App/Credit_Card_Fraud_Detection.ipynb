```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit= pd.read_csv("creditcard.csv", encoding= "latin1")
credit.head()
```

```python
credit.info()
```

```python
credit.shape
```

```python
credit['Class'].value_counts()
```

```python
credit.describe()
```

```python
legit= credit[credit.Class==0]
fraud= credit[credit.Class==1]
```

```python
fraud.head().value_counts()
```

```python
fraud.shape
```

```python
legit.shape
```

```python
legit_sample= legit.sample(n= 492)
credit= pd.concat([legit_sample,fraud],axis= 0)
```

```python
credit['Class'].value_counts()
```

```python
credit.groupby('Class').mean()
```

```python
X= credit.drop(columns= 'Class',axis= 1)
Y= credit['Class']
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size= 0.2,stratify= Y,random_state= 2)
```

```python
model= LogisticRegression()
model.fit(X_train, Y_train)
ypred= model.predict(X_test)
```

```python
accuracy_score(ypred,Y_test)
```
